```python
#Random Forest iris 데이터 분류

from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
#training과 test 데이터 설정
iris = load_iris()

#데이터의 시작부터 끝에서 29번까지 선택
x_train = iris.data[:-30]
y_train = iris.target[:-30]

#마지막 29번째 데이터부터 끝까지 선택
x_test = iris.data[-30:]
y_test = iris.target[-30:]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
#random forest 분류기 학습
rfc.fit(x_train, y_train)

#test data를 입력해 target data를 예측
prediction = rfc.predict(x_test)

#예측 결과 precision과 실제 test data의 target 을 비교
print(prediction == y_test)
```

```python
#Random Forest 분류기 성능 평가
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print("Accuracy is:", accuracy_score(prediction, y_test))
```

```python
#training, test 데이터 재생성
from sklearn.model_selection import train_test_split

x = iris.data
y = iris.target

#데이터를 무작위로 혼합한 후 8:2로 나누어 train과 test에 대입
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, Y_train)
```

```python
prediction_1 = clf.predict(X_test)

print("Accuracy is:", accuracy_score(prediction_1, Y_test))
```

```python
#붓꽃 종류 예측 결과 확인
arr = np.array([[5.1, 4.8, 5.6, 2.7]])

#1: setosa, 2: versicolor, 3: virginica
print("예측값=", clf.predict(arr))
```
